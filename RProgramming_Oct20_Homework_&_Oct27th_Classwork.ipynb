```python

# 2nd one
busiest_routes <- flights %>%
  group_by(origin, dest) %>%
  summarize(num_flights = n()) %>%
  arrange(desc(num_flights)) %>%
  filter(num_flights >= 1000)

print("Solution for Exercise 2:")
busiest_routes
```

```python
#3rd one

carrier_performance <- flights %>%
  group_by(carrier) %>%
  summarize(
    total_flights = n(),
    average_delay = mean(arr_delay, na.rm = TRUE),
    on_time_arrival_rate = mean(arr_delay <= 15, na.rm = TRUE),
    average_distance = mean(distance, na.rm = TRUE)
  ) %>%
  arrange(desc(on_time_arrival_rate))
```

```python
print("Solution for Exercise 3:")
print(carrier_performance)
```

```python

#4th one

monthly_trends <- flights %>%
  mutate(peak_season = ifelse(month %in% 6:8, "peak", "not_peak")) %>%
  group_by(month, peak_season) %>%
  summarize(
    total_flights = n(),
    average_distance = mean(distance, na.rm = TRUE),
    cancellation_rate = mean(is.na(dep_time)),
    average_delay = mean(arr_delay, na.rm = TRUE)
  )
```

```python
print("The answer for this is")
print(monthly_trends)
```

```python
#5th one

revenue_analysis <- flights %>%
  mutate(
    estimated_price = case_when(
      distance < 500 ~ 150,
      distance >= 500 & distance <= 1000 ~ 250,
      distance > 1000 ~ 400
    ),
    peak_season_premium = ifelse(month %in% 6:8, 1.20, 1),
    estimated_revenue = estimated_price * peak_season_premium
  ) %>%
  group_by(carrier) %>%
  summarize(
    total_revenue = sum(estimated_revenue, na.rm = TRUE)
  ) %>%
  arrange(desc(total_revenue))
```

```python

print("Solution for Exercise 5:")
print(revenue_analysis)
```

```python
#6th one

market_competition <- flights %>%
  group_by(origin, dest) %>%
  summarize(
    num_carriers = n_distinct(carrier),
    carrier_market_share = prop.table(table(carrier)),
    avg_price = mean(distance), # Placeholder, needs refinement
    on_time_performance = mean(arr_delay <= 15, na.rm = TRUE)
  ) %>%
  filter(num_carriers >= 3) %>%
  mutate(monopoly = ifelse(max(carrier_market_share) > 0.8, "Yes", "No"))

print("Solution for Exercise 6:")
market_competition
```
